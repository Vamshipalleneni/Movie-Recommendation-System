```python
# But we need the genres as list of stings but it is in string format
# we need the like ['Action','Adventure','Fantasy','Scifi']
```

```python
import ast
```

```python
def convert(obj):
  L= []
  for i in ast.literal_eval(obj):
    L.append(i["name"])
  return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head(1)
```

```python
movies.iloc[0].keywords
```

```python
# Same with keywords aswell

movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
movies.iloc[0]["keywords"]
```

```python
# need only the top3 cast

def convert3(obj):
  L= []
  count= 0
  for i in ast.literal_eval(obj):
    if count <3:
      L.append(i["name"])
      count= count+1
    else:
      break
  return L
```

```python
movies["cast"]= movies["cast"].apply(convert3)
```

```python
movies.iloc[0]["cast"]
```

```python
movies.iloc[0]["crew"]
```

```python
#need only the director name form the crew

def fetchdirector(obj):
  L= []
  for i in ast.literal_eval(obj):
    if i["job"]=="Director":
      L.append(i["name"])
      break
  return L
```

```python
movies["crew"]= movies["crew"].apply(fetchdirector)
```

```python
movies.iloc[0]["crew"]
```

```python
movies.head()
```

```python
# we need to convert the overview to list for that we will use the lambda function

movies["overview"]= movies["overview"].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
type(movies.iloc[0]["overview"])
```

```python
#Now remove whitespace between names

movies["genres"]= movies["genres"].apply(lambda x:[i.replace(" ","") for i in x])
movies["keywords"]= movies["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
movies["cast"]= movies["cast"].apply(lambda x:[i.replace(" ","") for i in x])
movies["crew"]= movies["crew"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies["tags"]= movies["overview"] + movies["genres"] + movies["keywords"] + movies["cast"] + movies["crew"]
```

```python
movies.head(1)
```

```python
new_df = movies[["movie_id","title","tags"]].copy()
```

```python
new_df.head(1)
```

```python
new_df.iloc[0]["tags"]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.iloc[0]["tags"]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
# We will perform Stemming
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  L= []
  for i in text.split():
    L.append(ps.stem(i))
  return " ".join(L)
```

```python
new_df["tags"]= new_df["tags"].apply(stem)
```

```python
new_df.iloc[0]['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words="english")
```

```python
vectors= cv.fit_transform(new_df["tags"]).toarray()
```

```python
vectors.shape
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
# now we will calculate the cosine similarity

from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity= cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
similarity
```

```python
similarity[0]
```

```python
new_df[new_df["title"]=="Batman Begins"].index[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse= True, key= lambda x:x[1])[1:6]
```

```python
def recommend(movie):
  movie_index = new_df[new_df["title"]== movie].index[0]
  distances= similarity[movie_index]
  movies_list= sorted(list(enumerate(distances)),reverse= True, key= lambda x:x[1])[1:6]
  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
